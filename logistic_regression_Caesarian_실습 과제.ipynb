```python
# logistic_regression_Caesarian_실습 과제

import tensorflow as tf
import numpy as np
tf.random.set_seed(5)
```

```python
# 데이터 불러오기
xy = np.loadtxt('caesarian.csv', delimiter=',', dtype=np.float32)
xy.shape     # (80, 6)

# 학습 데이터 분리 : 70%  , 56개

x_train = xy[:56,:-1]   # (56, 5)
y_train = xy[:56,[-1]]  # (56, 1)
print(x_train.shape)
print(y_train.shape)

# 검증 데이터 분리 : 30%  , 24개

x_test = xy[56:,:-1]
y_test = xy[56:,[-1]]
print(x_test.shape)      # (24, 5)
print(y_test.shape)      # (24, 1)
```

```python
# 변수 초기화 : weight, bias
# (m,n)  *  (n.l) = (m,l) : 행렬의 내적 곱셈 공식
#   X    *   W    =  Y
# (56,5) * (5,1) = (56, 1)
W = tf.Variable(tf.random.normal([5,1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
# W,b
```

```python
# 예측 함수 (hypothesis) : H(x) = sigmoid(tf.matmul(X,W) +  b)
# tf.sigmoid() : tf.div(1.,1. + tf.exp(-(tf.matmul(X,W) +  b)))
def hypothesis(X):
    return tf.sigmoid(tf.matmul(X,W) + b)  # 0과 1사이의 값이 출력
```

```python
# 비용 함수 : logloss, 2진 분류 모델

def cost_func():
    cost = -tf.reduce_mean(y_train*tf.math.log(hypothesis(x_train)) + (1 - y_train) * tf.math.log(1-hypothesis(x_train)))
    return cost
```

```python
# 경사 하강법
# learning_rate(학습율)을 0.01로 설정하여 optimizer 객체를 생성
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# 학습 시작
print('****** Start Learning !!')
for step in range(10001):
    # cost를 minimize한다
    optimizer.minimize(cost_func, var_list=[W,b])
    
    if step % 100 == 0:
        print('%04d'%step,'cost:[',cost_func().numpy(),']',
             ' W:',W.numpy(),' b:',b.numpy())
    
print('****** Start Finishing !!')
```

```python
# 회귀계수, weight과 bias 출력
print('weight:', W.numpy())
print('bias:', b.numpy())
```

```python
# 정확도 측정 : accuracy computation
def predict(X):
    return tf.cast(hypothesis(X) > 0.5, dtype=tf.float32)  # 예측값을 0과 1로 변환

# 검증 데이터를 사용하여 예측 (70% : 30%)
preds = predict(x_test)
accuracy = tf.reduce_mean(tf.cast(tf.equal(preds,y_test), dtype=tf.float32))

print('Accuracy:', accuracy.numpy())    # Accuracy: 0.45833334
```
