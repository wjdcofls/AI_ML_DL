```python
# 실습 과제 Boston

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
tf.random.set_seed(5)
```

```python
# 데이터 불러오기
xy = np.loadtxt('boston_train.csv', delimiter=',', skiprows=1, dtype=np.float32)

x_train = xy[:,:-1]     # x
y_train = xy[:,[-1]]    # y
print(x_train.shape, y_train.shape)  # (400, 9) (400, 1)
```

```python
# 변수 초기화 : weight, bias
# (m,n)  *  (n.l) = (m,l) : 행렬의 내적 곱셈 공식
#   X    *   W    =  Y
# (400, 9) * (9,1)  = (400,1)
W = tf.Variable(tf.random.normal([9,1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
# W,b
```

```python
# 예측 함수 (hypothesis) : H(x) = X1*W1 + X2*W2 + X3*W3 + b
def hypothesis(X):
    return tf.matmul(X,W) + b  # 내적 곱셈
```

```python
# 비용 함수 : (H(X) - Y)^2 의 평균
# tf.square() : 제곱
# tf.reduce_mean() : 평균

def cost_func():
    cost = tf.reduce_mean(tf.square(hypothesis(x_train) - y_train))
    return cost
```

```python
# 경사 하강법
# learning_rate(학습율)을 0.01로 설정하여 optimizer 객체를 생성

# optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.01)
# optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=0.01)   # 성능이 향상됨 _ AdamOptimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# 학습 시작
print('****** Start Learning !!')
for step in range(10001):
    # cost를 minimize한다
    optimizer.minimize(cost_func, var_list=[W,b])
    
    if step % 100 == 0:
        print('%04d'%step,'cost:[',cost_func().numpy(),']',
             ' W:',W.numpy(),' b:',b.numpy())
    
print('****** Start Finishing !!')
```

```python
# 회귀계수, weight과 bias 출력
print('weight:', W.numpy())
print('bias:', b.numpy())
```

```python
# 예측
print('***** Predict')

x_data = np.loadtxt('boston_train.csv', delimiter=',', skiprows=1, dtype=np.float32)
x_test = x_data[:,:-1]
y_test = x_data[:,[-1]]
                
print(hypothesis(x_test).numpy())
```

```python
# 평가 : RMSE
def get_rmse(y_test,preds):
    squared_error = 0
    for k,_ in enumerate(y_test):
        squared_error += (preds[k] - y_test[k])**2
    mse = squared_error/len(y_test)  
    rmse = np.sqrt(mse)
    return rmse[0]

# 학습한 데이터를 그대로 검증 데이터로 사용한 경우
x_test = x_data[:,:-1]
y_test = x_data[:,[-1]]

preds = hypothesis(x_test).numpy()
print('RMSE:', get_rmse(y_test,preds))   # RMSE: 2.4112918
```
