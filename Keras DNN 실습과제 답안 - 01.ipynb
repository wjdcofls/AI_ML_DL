## Keras DNN 실습과제 답안 - 01

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
# import csv
tf.random.set_seed(5)
np.random.seed(5)
```

```python
boston_train = np.loadtxt('boston_train.csv', delimiter=',', dtype=np.float32, skiprows=1)
boston_test = np.loadtxt('boston_test.csv', delimiter=',', dtype=np.float32, skiprows=1)

x_train = boston_train[:,:9]
y_train = boston_train[:,[-1]]
print(x_train.shape)   # (400,9)
print(y_train.shape)   # (400,1)
```

```python
# 모델 구현
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=20, input_shape=[9],activation=tf.nn.relu),
    tf.keras.layers.Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
# 학습
history = model.fit(x_train, y_train, epochs=700)
```

```python
# 시각화
epoch_count = range(1, len(history.history['loss']) + 1)
plt.plot(epoch_count, history.history['loss'], 'r-')
plt.legend(['Training Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
# 평가
x_test = boston_test[:,:9]
y_test = boston_test[:,[-1]]
mse = model.evaluate(x_test, y_test)
# rmse = np.sqrt(mse)
print('MSE:',mse)
```

## 모델 개선 : 3 layers

```python
tf.keras.backend.clear_session()
tf.random.set_seed(5)
np.random.seed(5) 
```

```python
# 모델 구현
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=20, input_shape=[9],activation=tf.nn.relu),
    tf.keras.layers.Dense(units=10, activation=tf.nn.relu),
    tf.keras.layers.Dense(units=1)    
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
# 학습
history = model.fit(x_train, y_train, epochs=700)
```

```python
# 시각화
epoch_count = range(1, len(history.history['loss']) + 1)
plt.plot(epoch_count, history.history['loss'], 'r-')
plt.legend(['Training Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
# 평가
x_test = boston_test[:,:9]
y_test = boston_test[:,[-1]]
mse = model.evaluate(x_test, y_test)
# rmse = np.sqrt(mse)
print('MSE:',mse)
```
