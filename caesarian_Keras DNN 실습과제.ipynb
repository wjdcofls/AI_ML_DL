```python
# Keras DNN 실습과제

import tensorflow as tf
import numpy as np
import csv
import matplotlib.pyplot as plt
tf.random.set_seed(5)

xy = np.loadtxt('caesarian.csv',delimiter=',',dtype=np.float32)

# train data set
x_data = xy[:56, :-1 ]
y_data = xy[:56, [-1] ]

x_train = np.array(x_data,dtype=np.float32)
y_train = np.array(y_data,dtype=np.float32)

x_test = xy[56:,:-1]
y_test = xy[56:,[-1]]

print(x_train.shape)
print(y_train.shape)
```

```python
model = tf.keras.Sequential([
    # (56,5) * (5,20) = (3,20) , Param : 120(w:500, b:100)
    tf.keras.layers.Dense(units=20, activation='relu', input_shape=(5,)),
    
    # (3,20) * (20,10) = (3,10) , Param : 210(w:5000, b:50)
    tf.keras.layers.Dense(units=10, activation='relu'),
    
    # (3,10) * (10,1) = (56,1) , Param : 11(w:25*5, b:5)
    tf.keras.layers.Dense(units=1)
    
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
             loss='mse') # 회귀모델
model.summary()
```

```python
# 학습
history = model.fit(x_train, y_train, epochs=700)
```

```python
## 시각화
#list(history.history['loss'])

epoch_count = range(1,len(history.history['loss']) + 1)
plt.plot(epoch_count, history.history['loss'],'b-')
plt.legend(['Training Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
# 예측
model.predict(x_train),y_train
```

```python
# 평가
model.evaluate(x_test,y_test)   # mse : 0.32566937804222107
```
